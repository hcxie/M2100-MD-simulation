```python
import numpy as np
import pandas as pd
import math as math
import matplotlib.pyplot as plt
import seaborn as sns
import math as math
%matplotlib inline
```

```python
dist_df=pd.DataFrame(columns=["AA results","CG results"])
```

```python
for i in [0,1,2,3,17,18]:
    filename="dist_c"+str(i+1)+"c"+str(i+2)+".txt"
    indexname="c"+str(i+1)+"c"+str(i+2)
    dist=np.loadtxt(filename, delimiter="\n", skiprows=1)   
    cdist=np.loadtxt("c"+filename, delimiter="\n", skiprows=1)
    dist_df.loc[indexname]=[dist,cdist]
```

```python
dist_df.shape
```

```python
dist_df_par=pd.DataFrame(columns=["AA mean","AA std","AA K","CG mean","CG std","CG k"])
```

```python
for i in [0,1,2,3,17,18]:
    filename="dist_c"+str(i+1)+"c"+str(i+2)+".txt"
    indexname="c"+str(i+1)+"c"+str(i+2)
    dist=np.loadtxt(filename, delimiter="\n", skiprows=1)   
    cdist=np.loadtxt("c"+filename, delimiter="\n", skiprows=1)
    aa_mean=np.mean(dist)
    aa_std=np.std(dist)
    aa_K=300/(2*aa_std**2)*1.38064852*(10**-23)/4184*(6.022*10**23)
    cg_mean=np.mean(cdist)
    cg_std=np.std(cdist)
    cg_K=300/(2*cg_std**2)*1.38064852*(10**-23)/4184*(6.022*10**23)
    dist_df_par.loc[indexname]=[aa_mean,aa_std,aa_K,cg_mean,cg_std,cg_K]
```

```python
dist_df_par
```

```python
def compareplot(distdata,i):
    plt.figure(figsize = [5, 5])
    sns.distplot(distdata.iloc[i,0],bins=20,kde=True,hist_kws = {'alpha' : 0.1})
    sns.distplot(distdata.iloc[i,1],bins=20,kde=True,hist_kws = {'alpha' : 0.1})

for i in range(0,dist_df.shape[0]):
    compareplot(dist_df,i)
```

```python
angl_df_par=pd.DataFrame(columns=["AA mean","AA std","AA K","CG mean","CG std","CG k"])
```

```python
for i in [0,1,2,4,16,17,18]:
    filename="angle_c"+str(i+1)+"c"+str(i+2)+"c"+str(i+3)+".txt"
    indexname="c"+str(i+1)+"c"+str(i+2)+"c"+str(i+3)
    angl=np.loadtxt(filename, delimiter="\n", skiprows=1)   
    cangl=np.loadtxt("c"+filename, delimiter="\n", skiprows=1)
    aa_mean=np.mean(angl)/math.pi*180
    aa_std=np.std(angl)
    aa_K=300/(2*aa_std**2)*1.38064852*(10**-23)/4184*(6.022*10**23)
    cg_mean=np.mean(cangl)/math.pi*180
    cg_std=np.std(cangl)
    cg_K=300/(2*cg_std**2)*1.38064852*(10**-23)/4184*(6.022*10**23)
    angl_df_par.loc[indexname]=[aa_mean,aa_std,aa_K,cg_mean,cg_std,cg_K]
```
